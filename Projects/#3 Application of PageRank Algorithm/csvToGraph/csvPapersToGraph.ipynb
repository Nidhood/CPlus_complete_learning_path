# Welcome to the program to convert a list of paper references into a reference graph!

### 1. First, we will read the CSV file into a DataFrame and look at the first few rows of the DataFrame:

```python
import pandas as pd
```

### 2. Read the CSV file into a DataFrame:

```python
df = pd.read_csv('research_papers_on_hand-drawn_circuit_recognizer.csv')
```

### 3. Look at the first few rows of the DataFrame:

```python
df.head()
```

### 4. Create a list of tuples from the DataFrame:

```python
tuplas = [(paper, paper_ref) for paper, paper_ref in zip(df['Paper'], df['Paper_Ref'])]
```

### 5. Look at the first few tuples:

```python
tuplas
```

### 6. Verify one of the Paper's names point to their references:

```python
for tupla in tuplas:
    if tupla[0] == 'Recognizing Hand-drawn Diagrams in Images. Ph.D. Thesis. Universit�t Mannheim':
        print(tupla)
```

### 7. Create a dictionary to assign a number to each paper name:

```python
nombre_a_numero = {}
contador = 0

for tupla in tuplas:
    nombre_papel = tupla[0]
    if nombre_papel not in nombre_a_numero:
        nombre_a_numero[nombre_papel] = contador
        contador += 1
```

### 8. Create a list to store the name of the paper and its assigned number:

```python
nuevas_tuplas = [(nombre_a_numero[tupla[0]], tupla[1]) for tupla in tuplas]
nombre_numero_lista = []

for nombre_papel, numero in nombre_a_numero.items():
    nombre_numero_lista.append((nombre_papel, numero))
```

### 9. Print the last 10 values:

```python
nuevas_tuplas[-10:]
```

### 10. Like in the previous step, we will create a dictionary to assign a number to each paper reference:

```python
for tupla in tuplas:
    nombre_papel = tupla[1]
    if nombre_papel not in nombre_a_numero:
        nombre_a_numero[nombre_papel] = contador
        contador += 1
```

### 11. Create a list to store the name of the paper and its assigned number:

```python
for nombre, numero in nombre_a_numero.items():
    nombre_numero_lista.append((nombre, numero))
```

### 13. Create a list to store the name of the paper and its assigned number:

```python
nuevas_tuplas_con_numeros = [(tupla[0], nombre_a_numero[tupla[1]]) for tupla in nuevas_tuplas]
```

### 14. Print the last 10 values:

```python
nuevas_tuplas_con_numeros[-10:]
```

### 15. Save the list of tuples to a text file:

```python
with open('/Users/nidhood/Desktop/carpeta sin título/New/C_complete_learning_path/Projects/#3 Application of PageRank Algorithm/data/reference_graph.txt', 'w') as file:
    # Itera sobre cada tupla en nuevas_tuplas_con_numeros y escribe en el archivo
    for tupla in nuevas_tuplas_con_numeros:
        # Escribir en el archivo en el formato deseado
        file.write(f"{tupla[0]}>{tupla[1]}\n")
```

### 16. Save the correspondence between the paper names and their assigned numbers to a text file:

```python
# Abre un archivo de texto para escribir
with open('number_name_equivalence.csv', 'w') as file:
    # Itera sobre cada tupla en nuevas_tuplas_con_numeros y escribe en el archivo
    for tupla in nombre_numero_lista:
        # Escribir en el archivo en el formato deseado
        file.write(f"{tupla[0]},{tupla[1]}\n")
```
